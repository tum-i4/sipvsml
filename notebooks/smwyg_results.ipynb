```python
def percentage(val):
    return f'{float(val)*100:.2f}'


def read_localization_results(csv_path):
    with open(csv_path) as inp:
        data = []
        headers = next(inp).strip().split(';')
        for line in inp:
            data.append(dict(zip(headers, line.strip().split(';'))))
    
    df = pd.DataFrame(data).set_index('obfuscation')
    
    for col in ['none', 'sc_guard', 'oh_verify', 'cfi_verify']:
        df[col] = df[col].map(percentage)
        
    for col in ['none_std', 'sc_std', 'oh_std', 'cfiv_std']:
        df[col] = df[col].map(lambda x: f'{float(x):.5f}')

    return df
```

```python
read_localization_results(data_dir / 'localization_simple_cov_original.csv')
```
